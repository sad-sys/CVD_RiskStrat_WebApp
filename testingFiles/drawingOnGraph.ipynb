```python
import numpy as np
import pandas as pd

heights = np.arange(140, 211)  # cm
population = np.exp(-0.5 * ((heights - 175) / 10) ** 2)

df = pd.DataFrame({
    "Height": heights,
    "Population": population
})

df.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(df["Height"], df["Population"])
plt.xlabel("Height (cm)")
plt.ylabel("Population")
plt.show()
```

```python
percentiles = [3, 50, 75]
labels = ["Lower Threshold", "Current Score", "Higher Threshold"]

cdf = np.cumsum(df["Population"])
cdf = cdf / cdf.iloc[-1]

plt.plot(df["Height"], df["Population"])

for p, label in zip(percentiles, labels):
    x = df.loc[cdf >= p/100, "Height"].iloc[0]
    plt.axvline(x, linestyle="--", label=label)

plt.xlabel("Height (cm)")
plt.ylabel("Population")
plt.legend()
plt.show()
```
