```python
import joblib
import numpy as np
import pandas as pd

# Load fitted pipeline components
imputer = joblib.load("/Users/sadiqkhawaja/Desktop/PhD/CVD_RiskStrat_WebApp-merge-interface2/model_files/imputers/model1_sociodemographics_rf (1).pkl")
scaler = joblib.load("/Users/sadiqkhawaja/Desktop/PhD/CVD_RiskStrat_WebApp-merge-interface2/model_files/scalers/scaler.pkl")
model = joblib.load("/Users/sadiqkhawaja/Desktop/PhD/CVD_RiskStrat_WebApp-merge-interface2/model_files/ML_models/MRMR_COX_Sociodemographics.pkl")

# Load example input
backend_values = {
    "category_Sociodemographics_ts_Length.of.time.at.current.address...Instance.0": 5.0,

    "category_Sociodemographics_ts_Type.of.accommodation.lived.in...Instance.0_1": 1,
    "category_Sociodemographics_ts_Type.of.accommodation.lived.in...Instance.0_4": 0,
    "category_Sociodemographics_ts_Type.of.accommodation.lived.in...Instance.0_neg7": 0,
    "category_Sociodemographics_ts_Type.of.accommodation.lived.in...Instance.0_3": 0,
    "category_Sociodemographics_ts_Type.of.accommodation.lived.in...Instance.0_5": 0,

    "Age.at.recruitment": 60.0,
}

sampleCols = list(imputer.feature_names_in_)
sample = pd.DataFrame(columns=sampleCols)

row = pd.Series(np.nan, index=sample.columns)

#NaNs instead
row = pd.Series(float('nan'), index=sample.columns)

# Step 3: Fill row
for key, value in backend_values.items():
    if key in row.index:
        row[key] = value
    else:
        print(f"⚠️ Missing column in sample: {key}")

# Step 4: Convert to DataFrame
final_input = pd.DataFrame([row])
# Raw input
X_raw = final_input.copy()
```

```python
import numpy as np

# Raw input row/dataframe with possible missing values
X_raw = final_input.copy()

# Original scikit-learn imputation
X_imputed_original = imputer.transform(X_raw)

print("Original imputed shape:", X_imputed_original.shape)
print(X_imputed_original[0])
```

```python
# Original scikit-learn RobustScaler transformation
X_scaled_original = scaler.transform(X_imputed_original)

print("Original scaled shape:", X_scaled_original.shape)
print(X_scaled_original[0])
```

```python
# Original fitted model prediction
risk_original = model.predict(X_scaled_original)

print("Original risk prediction:", risk_original[0])
```

```python
# Export the fitted random forests from IterativeImputer
exported_triplets = []

for triplet in imputer.imputation_sequence_:
    rf = triplet.estimator
    exported_forest = []

    for tree in rf.estimators_:
        exported_forest.append({
            "feature": tree.tree_.feature.copy(),
            "threshold": tree.tree_.threshold.copy(),
            "left": tree.tree_.children_left.copy(),
            "right": tree.tree_.children_right.copy(),
            "value": tree.tree_.value[:, 0, 0].copy(),
        })

    exported_triplets.append({
        "feat_idx": int(triplet.feat_idx),
        "neighbor_feat_idx": triplet.neighbor_feat_idx.copy(),
        "forest": exported_forest,
    })


def run_tree(x, tree):
    node = 0

    while tree["left"][node] != -1:
        feature_idx = tree["feature"][node]
        threshold = tree["threshold"][node]

        if x[feature_idx] <= threshold:
            node = tree["left"][node]
        else:
            node = tree["right"][node]

    return tree["value"][node]


def run_forest(x, forest):
    return np.mean([
        run_tree(x, tree)
        for tree in forest
    ])


def custom_impute(x, imputer, exported_triplets):
    x = np.asarray(x, dtype=float).reshape(1, -1)

    # Record which values were genuinely missing in the raw input
    missing_mask = np.isnan(x)

    # Reproduce IterativeImputer's initial filling step
    x_filled = imputer.initial_imputer_.transform(x)

    # Work with one row
    x_filled = np.asarray(x_filled[0], dtype=float)

    # Apply each fitted estimator in sklearn's stored order
    for triplet in exported_triplets:
        target_idx = triplet["feat_idx"]

        # Only update values that were missing in the original input
        if missing_mask[0, target_idx]:
            neighbor_indices = triplet["neighbor_feat_idx"]
            neighbor_values = x_filled[neighbor_indices]

            prediction = run_forest(
                neighbor_values,
                triplet["forest"],
            )

            # Match IterativeImputer clipping behaviour
            min_value = np.asarray(imputer._min_value)
            max_value = np.asarray(imputer._max_value)

            prediction = np.clip(
                prediction,
                min_value[target_idx],
                max_value[target_idx],
            )

            x_filled[target_idx] = prediction

    return x_filled
```

```python
raw_row = X_raw.iloc[0].to_numpy(dtype=float)

X_imputed_custom = custom_impute(
    raw_row,
    imputer,
    exported_triplets,
)

print("Maximum imputation difference:")
print(
    np.max(
        np.abs(X_imputed_custom - X_imputed_original[0])
    )
)
```

```python
# Export the fitted RobustScaler parameters
CENTER = np.asarray(scaler.center_, dtype=float)
SCALE = np.asarray(scaler.scale_, dtype=float)


def custom_scale(x):
    x = np.asarray(x, dtype=float)
    return (x - CENTER) / SCALE


X_scaled_custom = custom_scale(X_imputed_custom)

print("Maximum scaling difference:")
print(
    np.max(
        np.abs(X_scaled_custom - X_scaled_original[0])
    )
)
```

```python
# Export the fitted model coefficients
COEF = np.asarray(model.coef_, dtype=float)


def custom_predict(x_scaled):
    x_scaled = np.asarray(x_scaled, dtype=float)
    return np.dot(x_scaled, COEF)


risk_custom = custom_predict(X_scaled_custom)

print("Original risk:", risk_original[0])
print("Custom risk:", risk_custom)
print("Absolute prediction difference:", abs(risk_original[0] - risk_custom))
```

```python
import time
import pandas as pd

N_RUNS = 100


def benchmark(function, n_runs=N_RUNS):
    times = []

    for _ in range(n_runs):
        start = time.perf_counter()
        function()
        times.append(time.perf_counter() - start)

    times = np.asarray(times) * 1000  # milliseconds

    return {
        "Mean (ms)": times.mean(),
        "Median (ms)": np.median(times),
        "Min (ms)": times.min(),
        "Max (ms)": times.max(),
        "Std (ms)": times.std(),
    }


# Warm up the fitted estimators before measuring
imputer.transform(X_raw)
scaler.transform(X_imputed_original)
model.predict(X_scaled_original)

custom_impute(raw_row, imputer, exported_triplets)
custom_scale(X_imputed_custom)
custom_predict(X_scaled_custom)


timing_results = {
    "Original imputation": benchmark(
        lambda: imputer.transform(X_raw)
    ),
    "Original scaling": benchmark(
        lambda: scaler.transform(X_imputed_original)
    ),
    "Original prediction": benchmark(
        lambda: model.predict(X_scaled_original)
    ),
    "Custom imputation": benchmark(
        lambda: custom_impute(
            raw_row,
            imputer,
            exported_triplets,
        )
    ),
    "Custom scaling": benchmark(
        lambda: custom_scale(X_imputed_custom)
    ),
    "Custom prediction": benchmark(
        lambda: custom_predict(X_scaled_custom)
    ),
}

timing_df = pd.DataFrame(timing_results).T
timing_df
```
