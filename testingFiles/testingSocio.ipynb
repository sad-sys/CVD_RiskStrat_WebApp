```python
exported_triplets = []

for triplet in imputer.imputation_sequence_:

    rf = triplet.estimator

    exported_forest = []

    for tree in rf.estimators_:

        exported_forest.append({
            "feature": tree.tree_.feature.copy(),
            "threshold": tree.tree_.threshold.copy(),
            "left": tree.tree_.children_left.copy(),
            "right": tree.tree_.children_right.copy(),
            "value": tree.tree_.value[:, 0, 0].copy()
        })

    exported_triplets.append({
        "feat_idx": int(triplet.feat_idx),
        "neighbor_feat_idx": triplet.neighbor_feat_idx.copy(),
        "forest": exported_forest
    })

print("Triplets:", len(exported_triplets))
```

```python
def run_tree(x, tree):

    node = 0

    while tree["left"][node] != -1:

        f = tree["feature"][node]

        if x[f] <= tree["threshold"][node]:
            node = tree["left"][node]
        else:
            node = tree["right"][node]

    return tree["value"][node]
```

```python
def run_forest(x, forest):

    total = 0.0

    for tree in forest:
        total += run_tree(x, tree)

    return total / len(forest)
```

```python
def custom_impute(x, exported_triplets):

    x = np.asarray(x, dtype=float).copy()

    for triplet in exported_triplets:

        target = triplet["feat_idx"]

        if np.isnan(x[target]):

            neighbors = triplet["neighbor_feat_idx"]

            pred = run_forest(
                x[neighbors],
                triplet["forest"]
            )

            x[target] = pred

    return x
```

```python
row = final_input.iloc[0].values.astype(float)

manual = custom_impute(
    row,
    exported_triplets
)

sklearn = imputer.transform(final_input)[0]

print(
    np.max(
        np.abs(manual - sklearn)
    )
)
```

```python
import numpy as np

coef = model.coef_.tolist()
center = scaler.center_.tolist()
scale = scaler.scale_.tolist()

COEF = np.array(coef)
CENTER = np.array(center)
SCALE = np.array(scale)

def predict(X):
    X = np.asarray(X, dtype=float)
    X_scaled = (X - CENTER) / SCALE
    return np.dot(X_scaled, COEF)
```
